```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.dropna(inplace=True)
df
```

```python
df.columns.tolist()
```

```python
df['mpg'].describe()
```

```python
df['mpg'].plot(kind='hist')
plt.show()
```

```python
sns.displot(data=df, x='mpg')
```

```python
sns.boxplot(data=df,y='mpg')
```

```python
df["cylinders"].value_counts()
```

```python
df['cylinders'].value_counts().plot(kind='bar')
plt.show()
```

```python
sns.countplot(data=df, x='cylinders')
```

```python
df["origin"].value_counts()
```

```python
df["origin"].value_counts().plot(kind='bar')
plt.show()
```

```python
sns.countplot(data=df,x='origin')
```

```python
df.groupby("horsepower")["mpg"].mean()
```

```python
sns.catplot(data=df,x="horsepower",y="mpg",kind="bar")
```

```python
df.groupby("horsepower")["mpg"].mean().plot(kind='bar')
plt.show()
```

```python
df.groupby(["horsepower","model_year"])["mpg"].mean()
```

```python
df.groupby(["horsepower","model_year"])["mpg"].mean().plot(kind='bar')
plt.show()
```

```python
df.head()
```

```python
df.groupby(['model_year','origin'])['weight'].mean()
```

```python
df.groupby(['model_year','origin'])['weight'].mean().plot(kind='bar')
plt.show()
```

```python
df_origin_filter = df[df['origin']=='usa']
result=df_origin_filter.groupby(['model_year','origin'])['weight'].mean()
result.plot(kind='bar')
plt.show()
```
