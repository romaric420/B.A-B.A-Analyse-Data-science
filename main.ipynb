<h1> analyse Multivariee</h1>

```python
df = sns.load_dataset("tips")

df.head()
```

Discret / Discret

```python
pd.crosstab(df["sex"],df["smoker"],normalize=True)
```

```python
sns.heatmap(pd.crosstab(df["size"],df["day"]),annot=True,fmt="d")
```

Discret / continue

```python
df.groupby("sex")['tip'].mean()
```

```python
df.groupby("sex")['tip'].mean().describe().to_frame().T
```

```python
df.groupby("sex")['tip'].mean().plot(kind='bar')
plt.show()
```

```python
sns.catplot(data=df,x="sex",y="tip",kind="bar")
```

```python
sns.displot(data=df,x='tip',hue="sex")
```

```python
sns.catplot(data=df,x="sex",y="tip",kind="box")
plt.show()
```

Continue / continue

```python
plt.scatter(df['tip'],df['total_bill'])
plt.show()
```

```python
sns.scatterplot(data=df,x='tip',y='total_bill')
plt.show()
```
