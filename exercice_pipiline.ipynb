```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, PolynomialFeatures, LabelEncoder
from sklearn.feature_selection import SelectKBest, f_regression

# Charger les données
df = sns.load_dataset("mpg")

# 1. Éliminer les valeurs manquantes
df = df.dropna()

# 2. Séparer X/y
y = df['mpg']
X = df.drop('mpg', axis=1)

# 3. Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y)

# 4. Extraire le constructeur de "name"
constructors_connus = ['chevrolet', 'ford', 'plymouth', 'dodge', 'amc', 'pontiac', 
                       'buick', 'oldsmobile', 'mercury', 'chrysler', 'toyota', 
                       'datsun', 'volkswagen', 'honda', 'mazda', 'fiat', 'audi', 
                       'peugeot', 'renault', 'saab', 'volvo', 'bmw', 'subaru', 'nissan']

def get_constructor(name):
    name_lower = name.lower()
    for c in constructors_connus:
        if c in name_lower:
            return c
    return 'other'

X_train['constructor'] = X_train['name'].apply(get_constructor)
X_test['constructor'] = X_test['name'].apply(get_constructor)
X_train = X_train.drop('name', axis=1)
X_test = X_test.drop('name', axis=1)

# 5. Encoder origin (catégorie nominale -> one-hot simplifié en "usa" oui/non)
X_train['usa'] = (X_train['origin'] == 'usa').astype(int)
X_test['usa'] = (X_test['origin'] == 'usa').astype(int)
X_train = X_train.drop('origin', axis=1)
X_test = X_test.drop('origin', axis=1)

# Encoder constructor avec LabelEncoder
le = LabelEncoder()
X_train['constructor'] = le.fit_transform(X_train['constructor'])
X_test['constructor'] = le.transform(X_test['constructor'])

# 6. PolynomialFeatures sur acceleration, horsepower, weight
poly_cols = ['acceleration', 'horsepower', 'weight']
poly = PolynomialFeatures(degree=2, include_bias=False, interaction_only=True)

poly_train = poly.fit_transform(X_train[poly_cols])
poly_test = poly.transform(X_test[poly_cols])

# Noms des colonnes poly
poly_names = poly.get_feature_names_out(poly_cols)
poly_names = [n if ' ' not in n else n.replace(' ', '_') + '_poly' for n in poly_names]

# Créer DataFrames
poly_train_df = pd.DataFrame(poly_train, columns=poly_names, index=X_train.index)
poly_test_df = pd.DataFrame(poly_test, columns=poly_names, index=X_test.index)

# Remplacer les colonnes originales par les colonnes poly
X_train = X_train.drop(poly_cols, axis=1)
X_test = X_test.drop(poly_cols, axis=1)
X_train = pd.concat([X_train, poly_train_df], axis=1)
X_test = pd.concat([X_test, poly_test_df], axis=1)

# 7. Normalisation
minmax_cols = ['cylinders', 'model_year', 'usa', 'constructor']
standard_cols = [c for c in X_train.columns if c not in minmax_cols]

# MinMaxScaler
minmax = MinMaxScaler()
X_train[minmax_cols] = minmax.fit_transform(X_train[minmax_cols])
X_test[minmax_cols] = minmax.transform(X_test[minmax_cols])

# StandardScaler
standard = StandardScaler()
X_train[standard_cols] = standard.fit_transform(X_train[standard_cols])
X_test[standard_cols] = standard.transform(X_test[standard_cols])

# 8. SelectKBest k=10
selector = SelectKBest(score_func=f_regression, k=10)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Voir quelles colonnes ont été sélectionnées
selected_cols = X_train.columns[selector.get_support()]
print("Colonnes sélectionnées:", list(selected_cols))

# Convertir en DataFrame final
X_train_final = pd.DataFrame(X_train_selected, columns=selected_cols, index=X_train.index)
X_test_final = pd.DataFrame(X_test_selected, columns=selected_cols, index=X_test.index)

print(X_train_final.head())
```
